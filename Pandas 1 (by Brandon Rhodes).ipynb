```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as npy 
```

```python
x = pd.read_csv('bea.csv')
```

```python
# Ok to start off, we can call the beginning or end of our dataframe 'x' by writing x.head(), or x.tail() . 
# The default number of rows included is five unless a number is specified in the parentheses. x.head(10)

x.tail(3)
```

```python
# If we want to call a column, we can type either x.Year or x['Year'] If a column has any spaces in it for example, 
# "Quantity Footnotes" then the latter notation must be used. 

# NOTE: If an error message is given when trying to call a column, it is sometimes because the the name of the column
# in the original document includes unkown elements such as spaces. To see what python sees as the name of of a column
# just write df.columns.values 

x['Year']
```

```python
x.Year > 2004

# Once we have called a column, we have the ability to specify a condition for that column.
# For example, If we want to hilight just the years after 2004 we can write x.Year > 2004.
# What we recieve in the cell is a list of the indices that fulfill or do not fulfill the condition. 
```

```python
x[x.Year > 2004]

# If we write x[x.Year > 2004] we can highlight just the entries that are 'true', in other words
# the ones that fulfill the specified condition. We recieve a polished dataframe with just the 
# charcoal production after 2004. 
```

```python
x[x.Year > 1990].head(20)

# We can, of course, combine this concept with the .head() or .tail() concept to highlight just the beginning or end 
# of our list. 
```

```python
bra.head(7)[['Country', 'Year']]

# We can call 2 columns at once by writing the name of the dataset, follwed by double square brackets containing the
# desired columns in quotes. 
```

```python
x[(x.Year > 2007) | (x.Year < 1997)]

# If we want to exclude a certain year range, we can write the above to highlight every year except 1997-2008. 
```

```python
x[(x.Year > 2009) & (x.Year < 2013)] 

# We can use this same model to highlight a certain year range, in this case 2010-2012, using the '&' symbol
# instead of the pipe symbol '|'. 
```

```python
x[x.Year == 2014.0] 

# We can write the above to highlight only the charcoal production from 2014. Note that 2014.0 is not enclosed by 
# quotation marks. If we wanted to highlight a column that contained string values we would need quotation marks to 
# call the values. For example if we wanted to call the Country or Area 'Brazil', we would write:
# x[x.Country == 'Brazil']
```

```python
x[x.Country == 'Brazil']

# *NOTE* Sometimes making slight changes to the original csv file before it is uploaded can save you lots of time
# later. The original csv had "Country or Area" which led to errors when trying to highlight Brazil as it is below. 
# Changing 'Country or Area' to just 'Country' solved the problem. Subtle tweaks to csv files in Excell are usually 
# a good idea before uploading. 
```

```python
x[x.Country == 'Brazil'].sort_values(by=['Year'])

# We now know how to highlight an individual country and separate it from the rest of
# this mountain of data, which is awesome but we can organize it even further. 
# We can take all the brazil data and sort by whatever column we want by writing: .sort_values(by=[]) 
# The column we want to sort by goes inside the square brackets. 
```

```python
x[x.Country == 'Brazil'].sort_values(by='Year', ascending = False)

# Notice that in the above dataset the year values we have sorted are ascending instead of descending. We can fix this
# (if we want) by getting rid of the square brackets and writing ascending=False after our sorted value and separated 
# by a comma. 
```

```python
x[x.Country == 'Brazil'].sort_index().head(20)

# We can also sort by index by writing the above. 

```

```python
bra = x[x.Country == 'Brazil'].sort_values(by=['Year'])

# We can now simplify our brazil charcoal consumption organized by year by assigning it to a vairiable. 

```

```python
bra.head(250).plot(x='Year', y='Quantity')

# We can now use .plot to visualize the coal consumption of brazil by year. Notice that the lines in our graph
# are skewed by the fact that there are many different categories of coal consumption that are not distinguished. 
# We use the parentheses to specify what elements to compare against eachother as axes. 
```
